```python
text = soup.get_text()
```

```python
print(text)
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
```

```python
tokens = tokenizer.tokenize(text)
```

```python
words = [token.lower() for token in tokens]
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python
list(stop_words)[:10]
```

```python
no_stopwords_words = [word for word in words if word not in stop_words]
```

```python
count = Counter(no_stopwords_words)
```

```python
top_ten = count.most_common(10)
print(top_ten)
```
